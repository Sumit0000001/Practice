```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None
    
    def insert(self, data):
        node = Node(data)
        if self.root is None:
            self.root = node
        else:
            q = []
            q.append(self.root)
            while q:
                curr = q.pop(0)
                if curr.left is None:
                    curr.left = node
                    break
                else:
                    q.append(curr.left)
                if curr.right is None:
                    curr.right = node
                    break
                else:
                    q.append(curr.right)

    def delete_node(self, key):
        if self.root is None:
            return
        if self.root.left is None and self.root.right is None:
            if self.root.data == key:
                self.root = None
            else:
                return
        else:
            q = []
            q.append(self.root)
            while len(q) > 0:
                curr = q.pop(0)
                if curr.left:
                    q.append(curr.left)
                if curr.right:
                    q.append(curr.right)
                if curr.data == key:
                    x = curr
                y = curr
            if x:
                x.data = y.data
                if y.left:
                    y.data = y.left.data
                    y.left = None
                elif y.right:
                    y.data = y.right.data
                    y.right = None
                else:
                    y = None
                    
    def count_nodes(self, node):
        if node is None:
            return 0
        return 1 + self.count_nodes(node.left) + self.count_nodes(node.right)
    def count_leaf_nodes(self, node):
        if node is None:
            return 0
        if node.left is None and node.right is None:
            return 1
        else:
            return self.count_leaf_nodes(node.left) + self.count_leaf_nodes(node.right)
    def find_children(self, node):
        children = []
        if node.left:
            children.append(node.left.data)
        if node.right:
            children.append(node.right.data)
        return children

    def find_min(self,node):
        if node is None:
            return None
        res=[]
        temp=node
        min=None
        while res or temp:
            if temp:
                res.append(temp)
                temp=temp.left
            else:
                root=res.pop()
                if min is None or root.data<min.data:
                    min=root
                temp=root.right
        return min.data
            
    def find_max(self,node):
        if node is None:
            return None
        res=[]
        temp=node
        min=None
        while res or temp:
            if temp:
                res.append(temp)
                temp=temp.left
            else:
                root=res.pop()
                if min is None or root.data>min.data:
                    min=root
                temp=root.right
        return min.data
    def find_element(self,node,key):
        pass
    def diameter(self,node):
        if node is None:
            return 0
        left_height=self.height(node.left)
        right_height=self.height(node.right)
        left_dia=self.diameter(node.left)
        right_dia=self.diameter(node.right)
        return max(left_height+right_height+1,max(left_dia,right_dia))
    def height(self,node):
        if node is None:
            return 0
        else:
            left=self.height(node.left)
            right=self.height(node.right)
        return max(left,right)+1
    def level(self,node,key,level=1):
        pass
    def distance_between_two_nodes(self,node,key1,key2):
        pass
    def cousins(self,node,key):
        pass
    def siblings(self,node,key):
        pass
    def pre_order_traversal(self,node):
        pass
    def in_order_traversal(self,node):
        pass
    def post_order_traversal(self,node):
        pass
    def printTree(self,root):
        if root is None:
            return
        print(root.data, end=':')
        if root.left is not None:
            print('L',root.left.data,end=' ')
        if root.right is not None:
            print('R',root.right.data,end=' ')
        print()
        self.printTree(root.left)
        self.printTree(root.right)


bt=BinaryTree()
# root=bt.tree_input()
bt.insert(1)
bt.insert(2)
bt.insert(3)
bt.insert(4)
bt.insert(5)
bt.insert(6)
bt.insert(7)
bt.insert(8)
bt.insert(9)
bt.insert(10)
bt.insert(11)
bt.printTree(bt.root)
print('---------------------------------------------------------------')
# bt.delete_node(9)
# bt.printTree(bt.root)
#to count no of nodes in tree
print(bt.count_nodes(bt.root))

#to count no of leaf nodes
print(bt.count_leaf_nodes(bt.root))

#to find child of any nodes
node=bt.root.left.left
print('To find Children of',node.data, 'is:',bt.find_children(node))
print('To find min of Tree:',bt.find_min(bt.root))
print('To find max of Tree:',bt.find_max(bt.root))
print('The height of tree is:',bt.height(bt.root))
print("The Diameter of tree is:",bt.diameter(bt.root))
```

```python
def fact(n):
    memo=[0]*(n+1)
    memo[0]=1
    if n<0:
        return None
    if n==0:
        return 1
    for i in range(1,n+1):
        memo[i]= i* memo[i-1]
    return memo[n]
n=5
print(fact(n))
```

```python
import matplotlib.pyplot as plt

# Get the input data from the user
data = input("Enter the data as a space-separated list of numbers: ")
bin_size = int(input("Enter the bin size: "))

# Convert the input data to a list of integers
data = [int(x) for x in data.split()]

# Compute the histogram
histogram = {}
for x in data:
    bin_index = x // bin_size
    if bin_index in histogram:
        histogram[bin_index] += 1
    else:
        histogram[bin_index] = 1

# Plot the histogram
plt.bar(histogram.keys(), histogram.values(), width=1, align='edge')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.xticks(list(histogram.keys()))
plt.show()
```

```python
import numpy as np

# Get the input data from the user
data = input("Enter the data as a space-separated list of numbers: ")
bin_size = int(input("Enter the bin size: "))

# Convert the input data to a numpy array
data = np.array([int(x) for x in data.split()])

# Compute the histogram using numpy
hist, bins = np.histogram(data, bins=np.arange(min(data), max(data) + bin_size, bin_size))

# Print the histogram as a dictionary
print("Histogram:")
for i in range(len(hist)):
    print(f"{bins[i]} - {bins[i+1]-1}: {hist[i]}")
```

```python
# Get the input data from the user
data = input("Enter the data as a space-separated list of numbers: ")
bin_size = int(input("Enter the bin size: "))

# Convert the input data to a list of integers
data = [int(x) for x in data.split()]

# Compute the histogram
histogram = {}
for x in data:
    bin_index = x // bin_size
    if bin_index in histogram:
        histogram[bin_index] += 1
    else:
        histogram[bin_index] = 1

# Print the histogram
print("Histogram:")
for bin_index, count in histogram.items():
    bin_start = bin_index * bin_size
    bin_end = (bin_index + 1) * bin_size - 1
    bin_label = f"{bin_start} - {bin_end}"
    print(f"{bin_label}: {count}")
```

```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
class bt:
    def __init__(self):
        self.root=None
        
    def printTree(self,root):
        if root is None:
            return
        print(root.data,end=': ')
        if root.left is not None:
            print('L',root.left.data,end=' ')
        if root.right is not None:
            print('R',root.right.data,end=' ')
        print()
        printTree(root.left)
        printTree(root.right)
    def treeInput(self):
        rootdata=int(input())
        if rootdata== -1:
            return None
        root=BinaryTreeNode(rootdata)
        treeLeft=treeInput()
        treeRight=treeInput()
        root.left=treeLeft
        root.right=treeRight
        return root
    def count_Nodes(self, root):
        if root is None:
            return 0
        left_count= self.count_Nodes(root.left)
        right_count= self.count_Nodes(root.right)
        return left_count + right_count + 1
        
    def pre_order_traversal(self,root):
        if root:
            print(root.data,end=" ")
            self.pre_order_traversal(root.left)
            self.pre_order_traversal(root.right)

    def post_order_traversal(self,root):
        if root:
            self.post_order_traversal(root.left)
            self.post_order_traversal(root.right)
            print(root.data,end=" ")

    def in_order_traversal(self,root):
        if root:
            self.in_order_traversal(root.left)
            print(root.data,end=" ")
            self.in_order_traversal(root.right)

    def Max_of_tree(self,root):
        if root is None:
            return -1
        maxLeft = self.Max_of_tree(root.left)
        maxright=self.Max_of_tree(root.right)
        maximum= max(root.data,maxLeft,maxright)
        return maximum
    def height_tree(self,root):
        if root is None:
            return 0
        leftheight=self.height_tree(root.left)
        rightheight= self.height_tree(root.right)
        return max(leftheight,rightheight)+1

    def count_leaf(self,root):
        if root is None:
            return 0
        if root.left is None and root.right is None:
            return 1
        leftnodes = self.count_leaf(root.left)
        rightnodes= self.count_leaf(root.right)
        return leftnodes+rightnodes
    def isBalanced(self, root):
        if root is None:
            return True
        leftheight = self.height_tree(root.left)
        rightheight = self.height_tree(root.right)
        if abs(leftheight - rightheight) > 1:
            return False
        isleftbalanced = self.isBalanced(root.left)
        isrightbalanced = self.isBalanced(root.right)
        if isleftbalanced and isrightbalanced:
            return True
        else:
            return False


    def find_sibling(self, root, node):
        if root is None or node is None:
            return None
        if root.left == node:
            return root.right
        if root.right == node:
            return root.left
        sibling = self.find_sibling(root.left, node)
        if sibling is not None:
            return sibling
        sibling = self.find_sibling(root.right, node)
        return sibling

    def cousins(self, root, node):
        level = self.get_level(root, node)
        if level == 0:
            return []
        parent = self.get_parent(root, node)
        cousins = []
        self.get_cousins(parent, node, level - 1, cousins)
        return cousins

    def get_cousins(self, parent, node, level, cousins):
        if parent is None or level < 1:
            return
        if level == 1:
            if parent.left and parent.left != node:
                cousins.append(parent.left.data)
            if parent.right and parent.right != node:
                cousins.append(parent.right.data)
        else:
            self.get_cousins(parent.left, node, level - 1, cousins)
            self.get_cousins(parent.right, node, level - 1, cousins)

    def get_level(self, root, node, level=0):
        if root is None:
            return 0
        if root == node:
            return level
        l = self.get_level(root.left, node, level+1)
        if l > 0:
            return l
        return self.get_level(root.right, node, level+1)

    def get_parent(self, root, node):
        if root is None or root == node:
            return None
        if root.left == node or root.right == node:
            return root
        parent = self.get_parent(root.left, node)
        if parent is not None:
            return parent
        return self.get_parent(root.right, node)

bt = bt()
root = BinaryTreeNode(1)
root.left = BinaryTreeNode(2)
root.left.left = BinaryTreeNode(4)
root.left.right = BinaryTreeNode(5)
root.right = BinaryTreeNode(3)
root.right.left = BinaryTreeNode(6)
root.right.right = BinaryTreeNode(7)

# Print the tree
bt.printTree(root)

# Count the number of nodes
num_nodes = bt.count_Nodes(root)
print("Number of nodes:", num_nodes)

# Print the nodes using pre-order traversal
print("Nodes using pre-order traversal:", end=" ")
bt.pre_order_traversal(root)
print()

# Find the maximum value in the binary tree
max_val = bt.Max_of_tree(root)
print("Maximum value in the tree:", max_val)

# Find the height of the binary tree
height = bt.height_tree(root)
print("Height of the tree:", height)

# Count the number of leaf nodes in the binary tree
num_leaf_nodes = bt.count_leaf(root)
print("Number of leaf nodes:", num_leaf_nodes)

# Check whether the binary tree is balanced or not
is_balanced = bt.isBalanced(root)
if is_balanced:
    print("The binary tree is balanced")
else:
    print("The binary tree is not balanced")

# Find the cousins of node with data=5
node = root.left.right
cousins = bt.cousins(root, node)
print("Cousins of node with data=", node.data, ":", cousins)
```

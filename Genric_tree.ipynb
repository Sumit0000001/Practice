```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.children = []

    def insert_node(self, parent_val, new_node_val):
        queue = [self]
        while queue:
            curr_node = queue.pop(0)
            if curr_node.val == parent_val:
                curr_node.children.append(TreeNode(new_node_val))
                return
            queue.extend(curr_node.children)

    def buildTree():
        root_val = int(input())
        root = TreeNode(root_val)
        num_children = int(input())
        for i in range(num_children):
            child = TreeNode.buildTree()
            root.children.append(child)
        return root

    def countLeaves(root):
        if not root:
            return 0
        if not root.children:
            return 1
        count = 0
        for child in root.children:
            count += TreeNode.countLeaves(child)
        return count
    
    def height(root):
        if not root:
            return 0
        if not root.children:
            return 1
        max_height = 0
        for child in root.children:
            child_height = TreeNode.height(child)
            if child_height > max_height:
                max_height = child_height
        return max_height + 1

    def print_tree(self, level=0):
        if self:
            print("|-- "*level,str(self.val))
            for child in self.children:
                child.print_tree(level+1)
                
    def max_child_sum_node(self):
        self.max_node = None
        self.max_sum = float('-inf')
        self._dfs(self)
        return self.max_node

    def _dfs(self, node):
        child_sum = sum(child.val for child in node.children)
        if child_sum > self.max_sum:
            self.max_sum = child_sum
            self.max_node = node
        for child in node.children:
            self._dfs(child)
            
    def sumTree(self):
        sum = self.val
        for child in self.children:
            sum += child.sumTree()
        return sum

root = TreeNode(1)
root.insert_node(1, 2)
root.insert_node(1, 3)
root.insert_node(2, 4)
root.insert_node(2, 5)
root.insert_node(3, 6)

root.print_tree()
print("Number of leaves:", TreeNode.countLeaves(root))
print("Height of tree:", TreeNode.height(root))
max_node = root.max_child_sum_node()
print("Node with maximum child sum:", max_node.val)
print(root.sumTree())
```

```python
'''
Climbing Stairs:
You are climbing a staircase that has n steps. 
You can take either 1 or 2 steps at a time. 
How many distinct ways are there to climb to the top of the staircase?

#recursive
def climbStairs(n):
    if n == 0 or n == 1:
        return 1
    return climbStairs(n-1) + climbStairs(n-2)
    
#memoization
def climbStairs(n, memo={}):
    if n == 0 or n == 1:
        return 1
    if n in memo:
        return memo[n]
    memo[n] = climbStairs(n-1, memo) + climbStairs(n-2, memo)
    return memo[n]
    
#iteration
def climbStairs(n):
    if n == 0 or n == 1:
        return 1
    a, b = 1, 1
    for i in range(2, n+1):
        c = a + b
        a, b = b, c
    return c
'''
#Tabulation
def climbStairs(n):
    if n == 0 or n == 1:
        return 1
    dp = [0] * (n+1)
    dp[0] = dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
print(climbStairs(5))

```

```python
'''
Coin Change:
Given a set of coins and a target amount, 
find the minimum number of coins needed to make up the target amount. 
You can assume that you have an unlimited supply of each type of coin.

#recursion
def coin_change(coins, amount):
    if amount == 0:
        return 0
    min_coins = float('inf')
    for coin in coins:
        if amount - coin >= 0:
            min_coins = min(min_coins, coin_change(coins, amount - coin))
    return min_coins + 1


#memoization
def coin_change(coins, amount, memo={}):
    if amount in memo:
        return memo[amount]
    if amount == 0:
        return 0
    min_coins = float('inf')
    for coin in coins:
        if amount - coin >= 0:
            min_coins = min(min_coins, coin_change(coins, amount - coin, memo))
    memo[amount] = min_coins + 1
    return memo[amount]

#iteration
def coin_change(coins, amount):
    queue = [(amount, 0)]
    min_coins = float('inf')
    while queue:
        curr_amount, curr_coins = queue.pop(0)
        if curr_amount == 0:
            min_coins = min(min_coins, curr_coins)
        elif curr_amount > 0:
            for coin in coins:
                queue.append((curr_amount - coin, curr_coins + 1))
    return min_coins
'''
#tabulation
def coin_change(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
print(coin_change(coins, amount))
```

```python
'''
Longest Increasing Subsequence:
Given an array of integers, 
find the length of the longest increasing subsequence. 
A subsequence is a sequence that can be derived from an array by deleting 
some or no elements without changing the order of the remaining elements. 
For example, [3, 6, 2, 7] has a longest increasing subsequence of [3, 6, 7], with a length of 3.

#recursive
def lis(nums):
    def lis_helper(i, prev):
        if i == len(nums):
            return 0
        taken = 0
        if nums[i] > prev:
            taken = 1 + lis_helper(i+1, nums[i])
        not_taken = lis_helper(i+1, prev)
        return max(taken, not_taken)
    
    return lis_helper(0, float('-inf'))

#memoization
def lis(nums):
    memo = {}
    def lis_helper(i, prev):
        if i == len(nums):
            return 0
        if (i, prev) in memo:
            return memo[(i, prev)]
        taken = 0
        if nums[i] > prev:
            taken = 1 + lis_helper(i+1, nums[i])
        not_taken = lis_helper(i+1, prev)
        memo[(i, prev)] = max(taken, not_taken)
        return memo[(i, prev)]
    
    return lis_helper(0, float('-inf'))

#iteration
def lis(nums):
    n = len(nums)
    dp = [1] * n
    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
'''
#tabulation
def longest_increasing_subsequence(nums):
    n = len(nums)
    dp = [1] * n

    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)


# example runner code
nums = [10, 9, 2, 5, 3, 7, 101, 18]
print(longest_increasing_subsequence(nums)) 
print(lis(nums))
```

```python
'''
Knapsack Problem:
Given a set of items, each with a weight and a value, 
determine the number of each item to include in a collection 
so that the total weight is less than or equal to a given limit and 
the total value is as large as possible. You can assume that 
there is an unlimited supply of each item.
'''
#Tabulation
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0 for j in range(capacity+1)] for i in range(n+1)]
    
    for i in range(1, n+1):
        for j in range(1, capacity+1):
            if weights[i-1] > j:
                dp[i][j] = dp[i-1][j]
            else:
                dp[i][j] = max(dp[i-1][j], values[i-1] + dp[i][j-weights[i-1]])
    
    return dp[n][capacity]
weights = [1, 2, 3, 4, 5]
values = [1, 4, 5, 7, 10]
capacity = 8

result = knapsack(weights, values, capacity)

print(result)
```

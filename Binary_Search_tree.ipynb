```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None
    
    def insert(self, data):
        node = Node(data)
        if self.root is None:
            self.root = node
        else:
            q = []
            q.append(self.root)
            while q:
                curr = q.pop(0)
                if curr.left is None:
                    curr.left = node
                    break
                else:
                    q.append(curr.left)
                if curr.right is None:
                    curr.right = node
                    break
                else:
                    q.append(curr.right)

    def delete_node(self, key):
        if self.root is None:
            return
        if self.root.left is None and self.root.right is None:
            if self.root.data == key:
                self.root = None
            else:
                return
        else:
            q = []
            q.append(self.root)
            while len(q) > 0:
                curr = q.pop(0)
                if curr.left:
                    q.append(curr.left)
                if curr.right:
                    q.append(curr.right)
                if curr.data == key:
                    x = curr
                y = curr
            if x:
                x.data = y.data
                if y.left:
                    y.data = y.left.data
                    y.left = None
                elif y.right:
                    y.data = y.right.data
                    y.right = None
                else:
                    y = None
                    
    def count_nodes(self, node):
        if node is None:
            return 0
        return 1 + self.count_nodes(node.left) + self.count_nodes(node.right)
    def count_leaf_nodes(self, node):
        if node is None:
            return 0
        if node.left is None and node.right is None:
            return 1
        else:
            return self.count_leaf_nodes(node.left) + self.count_leaf_nodes(node.right)
    def find_children(self, node):
        children = []
        if node.left:
            children.append(node.left.data)
        if node.right:
            children.append(node.right.data)
        return children
    def find_parent(self,root, child):
        if root is None:
            return None
        if root.data==child:
            return
        if root.left.data == child or root.right.data == child:
            return root
        left = self.find_parent(root.left, child)
        if left:
            return left
        right = self.find_parent(root.right, child)
        if right:
            return right
        return None

    
    def find_min(self, node):
        if node is None:
            return None
        while node.left:
            node = node.left
        return node.data
    
    def find_max(self, node):
        if node is None:
            return None
        while node.right:
            node = node.right
        return node.data
    
    def find_element(self, node, key):
        if node is None:
            return False
        if node.data == key:
            return True
        if key < node.data:
            return self.find_element(node.left, key)
        else:
            return self.find_element(node.right, key)
    
    def preorder_traversal(self, node):
        if node:
            print(node.data, end=' ')
            self.preorder_traversal(node.left)
            self.preorder_traversal(node.right)
    
    def postorder_traversal(self, node):
        if node:
            self.postorder_traversal(node.left)
            self.postorder_traversal(node.right)
            print(node.data, end=' ')
    
    def inorder_traversal(self, node):
        if node:
            self.inorder_traversal(node.left)
            print(node.data, end=' ')
            self.inorder_traversal(node.right)
            
    def printTree(self,root):
        if root is None:
            return
        print(root.data,end=': ')
        if root.left is not None:
            print('L',root.left.data,end=' ')
        if root.right is not None:
            print('R',root.right.data,end=' ')
        print()
        self.printTree(root.left)
        self.printTree(root.right)

    
    def diameter(self, node):
        if node is None:
            return 0
        
        left_height = self.height(node.left)
        right_height = self.height(node.right)
        
        left_diameter = self.diameter(node.left)
        right_diameter = self.diameter(node.right)
        
        return max(left_height + right_height + 1, max(left_diameter, right_diameter))
    
    def height(self, node):
        if node is None:
            return 0
        else:
            left_height = self.height(node.left)
            right_height = self.height(node.right)
            return max(left_height, right_height) + 1
    
    def level(self, node, key, level=1):
        if node is None:
            return 0
        if node.data == key:
            return level
        left_level = self.level(node.left, key, level+1)
        if left_level:
            return left_level
        right_level = self.level(node.right, key, level+1)
        return right_level
    
    def distance(self, node, key1, key2):
        if node is None:
            return 0
        if node.data == key1:
            return self.level(node, key2)
        if node.data == key2:
            return self.level(node, key1)
        left_distance = self.distance(node.left, key1, key2)
        right_distance = self.distance(node.right, key1, key2)
        if left_distance != 0 and right_distance != 0:
            return left_distance + right_distance - 2*self.level(node, key1) 
        if left_distance != 0:
            return left_distance
        else:
            return right_distance
    
    def cousins(self, node, key):
        if node is None:
            return []
        
        parent = None
        current_level = [node]
        while current_level:
            next_level = []
            for n in current_level:
                if n.left and n.left.data == key:
                    parent = n
                elif n.right and n.right.data == key:
                    parent = n
                else:
                    if n.left:
                        next_level.append(n.left)
                    if n.right:
                        next_level.append(n.right)
            if parent:
                return [n.data for n in next_level if n != parent.left and n != parent.right]
            current_level = next_level
        return []
    
    def siblings(self, node, key):
        if node is None:
            return []
        
        parent = None
        current_level = [node]
        while current_level:
            next_level = []
            for n in current_level:
                if n.left and n.left.data == key:
                    parent = n
                elif n.right and n.right.data == key:
                    parent = n
                else:
                    if n.left:
                        next_level.append(n.left)
                    if n.right:
                        next_level.append(n.right)
            if parent:
                if parent.left and parent.right:
                    return [parent.left.data, parent.right.data]
                else:
                    return []
            current_level = next_level
        return []

# create an instance of BinaryTree
b = BinaryTree()

# insert nodes into the binary tree
b.insert(10)
b.insert(3)
b.insert(1)
b.insert(5)
b.insert(7)
b.insert(17)
b.insert(21)
b.insert(12)
b.insert(13)

# print the binary tree
b.printTree(b.root)
print('pre_order_traversal')
# perform a preorder traversal of the binary tree
b.preorder_traversal(b.root)
print()
print('post_order_traversal')

# perform a postorder traversal of the binary tree
b.postorder_traversal(b.root)
print()
print('in_order_traversal')

# perform an inorder traversal of the binary tree
b.inorder_traversal(b.root)
print()

# delete a node from the binary tree
b.delete_node(5)

# print the binary tree after node deletion
b.printTree(b.root)
#to print total no of nodes in bst
print("Total number of nodes in binary search tree:", b.count_nodes(b.root))

#to count leaf nodes in binary tree
print("Number of leaf nodes in binary tree:", b.count_leaf_nodes(b.root))

# find the children of a parent node
parent_node = b.root.left # assume the parent node is the left child of the root
children = b.find_children(parent_node)
print("Children of parent node {}: {}".format(parent_node.data, children))

node_value = 10
parent = b.find_parent(b.root, node_value)
if parent:
    print("Parent of node", node_value, "in binary tree:", parent.data)
else:
    print("Node", node_value, "has no parent in binary tree")
print()

# find the minimum value in the binary tree
min_val = b.find_min(b.root)
print("Minimum value in binary tree:", min_val)

# find the maximum value in the binary tree
max_val = b.find_max(b.root)
print("Maximum value in binary tree:", max_val)

# find whether a value exists in the binary tree
key = 12
if b.find_element(b.root, key):
    print("Element", key, "exists in binary tree")
else:
    print("Element", key, "does not exist in binary tree")
# calculate and print the diameter of the binary tree
print("Diameter of binary tree:", b.diameter(b.root))
print()

# calculate and print the height of the binary tree
print("Height of binary tree:", b.height(b.root))
print()

# calculate and print the level of a node in the binary tree
node_value = 7
print("Level of node", node_value, "in binary tree:", b.level(b.root, node_value))
print()

# calculate and print the shortest distance between two nodes in the binary tree
node1_value = 13
node2_value = 21
print("Shortest distance between nodes", node1_value, "and", node2_value, "in binary tree:", b.distance(b.root, node1_value, node2_value))
print()

# calculate and print the cousins of a node in the binary tree
node_value = 13
print("Cousins of node", node_value, "in binary tree:", b.cousins(b.root, node_value))
print()

# calculate and print the siblings of a node in the binary tree
node_value = 21
print("Siblings of node", node_value, "in binary tree:", b.siblings(b.root, node_value))
print()
```
